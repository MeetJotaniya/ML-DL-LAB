## Simple Linear Regression

## Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

## Step 2. Import the dataset

```python
df = pd.read_csv("50_Startups.csv")
```

```python
df
```

## Step 3 . Check the State Column

```python
df.value_counts('State')
```

## Step 4 . Splitting dataset in to input and output

```python
X = df.drop('Profit',axis=1)
y = df['Profit']

X
```

## Step 5 . Convert state Column into Numeric Column

## Step 5.1 .  Perform Transformation

```python
x1 = pd.get_dummies(X,columns=['State'],drop_first=True)
# drop_first will remove one col from which are generated from get dummies and it can be acceseed by other two column.
```

```python
x1
```

## Step 6 . Dummy variable trap

```python
# Already Performed using | drop_first =True
```

## Step 7 Splitting dataset in to Train and Test

```python
from sklearn.model_selection import train_test_split

x1_train, x1_test, y_train, y_test = train_test_split(x1,y,test_size=0.2,random_state=42)
```

## Step 8  Import LinearRegression model from linear_model family 

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

## Step 9  Fit the data

```python
model.fit(x1_train, y_train)
```

## Step 10  Predict the data

```python
y_pred = model.predict(x1_test)
```

## Step 11  Display Result

```python
# y_test and y_predict
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x1_test['Administration'],y_test)
plt.scatter(x1_test['Administration'],y_pred)
plt.show()
```

## RSS

```python
np.sum((y_test.values - y_pred) ** 2) 
```

```python
len(y_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test.values, y_pred) * len(y_pred)
```

```python
len(y_pred)
```

## R Square

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred)
```

## Now use Polynomial Regression on Position_Salaries dataset

```python
df2 = pd.read_csv("Position_Salaries.csv")
```

```python
df2

plt.scatter(df2['Level'],df2['Salary'])
plt.show()
```

```python
x2 = df2.drop(['Salary','Position'],axis=1)
y2 = df2['Salary']
x2
```

```python
x2_train, x2_test, y2_train, y2_test = train_test_split(x2, y2, random_state=42, test_size=0.3)
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=2)
```

```python
x3 = poly.fit_transform(x2_train)
x3
```

```python
model1 = LinearRegression()
```

```python
model1.fit(x3, y2_train)
```

```python
y_poly_pred = model1.predict(poly.fit_transform(x2_test))
y_poly_pred
```

```python
plt.scatter(x2_test['Level'],y2_test)
plt.scatter(x2_test['Level'],y_poly_pred)
plt.show()
```
